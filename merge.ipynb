```python
df_original_import_location = 'datasets/emoevent/raw/full_subset/emoevent_crowdsourced_400.csv'
df_fgcu_import_location = 'datasets/emoevent/raw/full_subset/fgcu_labels.csv'
df_ai_import_location = 'datasets/emoevent/raw/full_subset/ai_labeled/emoevents_ai_labeled_400.csv'
```

```python
import pandas as pd

df_original = pd.read_csv(df_original_import_location)
df_original
```

```python
df_original = df_original.rename({'Unnamed: 0': 'original index', 'labels': 'crowdsourced'}, axis=1)

# Move crowdsourced labels to after the text record
crowdsourced_labels = df_original.pop('crowdsourced')
df_original.insert(len(df_original.columns), 'crowdsourced', crowdsourced_labels)

df_original
```

```python
df_fgcu = pd.read_csv(df_fgcu_import_location)
df_fgcu
```

```python
df_ai = pd.read_csv(df_ai_import_location)
df_ai
```

```python
df_all = pd.concat([df_original, df_fgcu.drop(['context','crowdsourced','text'], axis=1), df_ai.drop(['original index','context','crowdsourced','text'], axis=1)], axis=1)
df_all
```

```python
df_all.to_csv('datasets/goemotions/ekman_single/full_subset/emoevent_400_all_labels.csv', index=False)
```
