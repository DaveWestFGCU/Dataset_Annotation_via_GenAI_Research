```python
import pandas as pd

import_location = './goemotions_labels.csv'
df = pd.read_csv(import_location)

df = df.rename(columns={'crowdsourced (original)': 'crowdsourced'})
df
```

```python
from sklearn.metrics import cohen_kappa_score

cohen_kappa_score(df['crowdsourced'], df['FGCU Combined'])
```

```python
cohen_kappa_score(df['crowdsourced'], df['AI Combined'])
```

```python
cohen_kappa_score(df['FGCU Combined'], df['AI Combined'])
```
